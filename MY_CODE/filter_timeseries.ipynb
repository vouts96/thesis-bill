# Notebook for filtering timeseries by year and ID

```python
import pandas as pd

def filter_timeseries(year, month, day, id, csv):
    ''' Gets as input a csv in a specific format (deepTSF long multiseries)
        and saves result to another csv filter by year or month or day or id or all the combinations 

    Parameters
    ----------
    year : int
        the number of the year we want to filter or None
    
    id : str
        the id of the timeseries we want to filter or None

    csv : str
        the path (absolute or relative) of the csv whose format we want to be modified

    Returns
    -------
    returns a dataframe filtered by year or month or day or id or all the combinations 
    '''

    # Read csv
    df = pd.read_csv(csv, parse_dates=['Date'])

    # Filter by ID if not None
    if(id != None):
        df = df.loc[df['ID'] == id].reset_index(drop=True)
        if(df.empty):
            return df
        df = df.drop('Unnamed: 0', axis=1)

    df.to_csv('../EXCEL_TO_CSV/' + id + '.csv')

    return df

filter_timeseries(0,0,0,'ALOYMINION','../EXCEL_TO_CSV/all_data_changed_format.csv')
```
