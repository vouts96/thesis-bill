# Notebook for adjusting the format of the data

```python
import pandas as pd


def change_format(csv):
    ''' Gets as input a csv in a specific format (rows present time and columns present areas)
        and returns a dataframe in a wanted format (rows present areas and columns present time)

    Parameters
    ----------
    csv : str
        the path (absolute or relative) of the csv whose format we want to be modified

    Returns
    -------
    returns a dataframe with the wanted format  
    '''

    df_before_modify = pd.read_csv(csv)
    cols = df_before_modify.columns[2:46]    # keep the columns that have names of cities/areas
    cols_dict = dict(zip(cols, range(len(cols))))   # create a dictionary with an index for each city/area as a value

    # create a new dataframe with the wanted columns
    d = {'Day': [], 'ID': [], 'Timeseries ID': [], 'Source': [], '00:00:00': [], '01:00:00': [],'02:00:00': [], '03:00:00': [], '04:00:00': [], '05:00:00': [], '06:00:00': [], '07:00:00': [], '08:00:00': [], '09:00:00': [], '10:00:00': [], '11:00:00': [], '12:00:00': [], '13:00:00': [], '14:00:00': [], '15:00:00': [], '16:00:00': [], '17:00:00': [], '18:00:00': [], '19:00:00': [], '20:00:00': [], '21:00:00': [], '22:00:00': [], '23:00:00': []}
    df = pd.DataFrame(data=d)
    df['Source'] = df['Source'].astype(str)
    df['ID'] = df['ID'].astype(int)
    df['Timeseries ID'] = df['ID'].astype(int)

    # read every 24 rows from the unmodified dataframe
    i = 0
    while i <= 100:# df_before_modify.shape[0]-24:


        temp_df = df_before_modify.iloc[i:i+24]
        temp_date = temp_df['DateTime'][i].split()[0]   # save the current date into a variable
        temp_df = temp_df.drop(columns=temp_df.columns[:2]) # drop the two first columns
        temp_df = temp_df.drop(columns=temp_df.columns[44:]) # drop the two first columns
        
        temp_df = temp_df.transpose()

        # iterate inside the temporary 24 row dataframe and append each column of it to each row
        # of the new dataframe 
        for j in range(i, i+24, 1):
            df.at[j,'Day'] = temp_date
            df.at[j,'Source'] = temp_df.index[j%24]
            df.at[j,'ID'] = cols_dict[temp_df.index[j%24]]
            df.at[j,'Timeseries ID'] = cols_dict[temp_df.index[j%24]]
            
            for z in range(0, 24, 1):
                df.iloc[j, z+4] = temp_df.iloc[j%24, z]


        i = i + 24
  
    

    return df


change_format('../EXCEL_TO_CSV/all_data.csv')
```
