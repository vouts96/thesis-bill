### Description of the dataframe

```python
import pandas as pd

def describe_df(csv):
    """Gets the relative or absolute path of a csv and calls the
        describe function of a dataframe

    Parameters
    ----------
    csv : str
        The relative or absolute path of the csv file

    Returns
    -------
    DataFrame
        A dataframe that contains trivial statistical info of our data 
    """

    df = pd.read_csv(csv)
    df = df.drop(columns=['Unnamed: 0'])    # Drops the index column of our data
    return df.describe()
    

describe_df('../EXCEL_TO_CSV/all_data.csv')
```

### Basic info of the dataframe

```python
import pandas as pd

def info_df(csv):
    """Gets the relative or absolute path of a csv and calls the
        info function of a dataframe

    Parameters
    ----------
    csv : str
        The relative or absolute path of the csv file

    Returns
    -------
    DataFrame
        A dataframe that contains info of our data 
    """

    df = pd.read_csv(csv)
    df = df.drop(columns=['Unnamed: 0'])    # Drops the index column of our data
    return df.info()
    

info_df('../EXCEL_TO_CSV/all_data.csv')
```

### Sum of Null Values of the dataset per region

```python
import pandas as pd

def describe_df(csv):
    """Gets the relative or absolute path of a csv and calls the
        isnull function of a dataframe in order to see how many null values there
        are in the dataset

    Parameters
    ----------
    csv : str
        The relative or absolute path of the csv file

    Returns
    -------
    DataFrame
        A dataframe that contains sum of null values of our data 
    """

    df = pd.read_csv(csv)
    df = df.drop(columns=['Unnamed: 0'])    # Drops the index column of our data
    return df.isnull().sum()
    

describe_df('../EXCEL_TO_CSV/all_data.csv')
```

From the output above we can se that there are some missing values in the dataset per station. Most stations having just 96 missing values which is not a significant problem 
because 96/24 = 4 days in total. However there are some stations where there are more than 10000 missing values like ELPE-HAR and TRIKALA where this might have a significant impact
on our dataset.

### Plotting 1H data based on region

```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_region(csv, region):
    """Gets a csv of data and a wanted region that you want to plot
        from the csv 
    
    Parameters
    ----------
    csv : str
        The absolute or relative path of the csv file

    Returns
    -------
    None
    """
    df = pd.read_csv(csv)
    df.plot(x='DateTime', y=region, kind='line', figsize=(15,10))
    

plot_region('../EXCEL_TO_CSV/all_data.csv', 'TRIKALA')
plot_region('../EXCEL_TO_CSV/all_data.csv', 'SERRES')
plot_region('../EXCEL_TO_CSV/all_data.csv', 'THESSALONIKI')
plot_region('../EXCEL_TO_CSV/all_data.csv', 'KILKIS')
plot_region('../EXCEL_TO_CSV/all_data.csv', 'KATERINI')
```

From the plots above we can see that there is a difference in the scale of the data. For example THESSALONIKI which has only 96 missing values, in plot it seems like there are no values until
07-08-2017 which is not true. If someone examines the data for THESSALONIKI, he will notice that there is a difference in scale.

### Plotting 1H Rescaled Data

```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_region_rescaled(csv, region):
    """Gets a csv of data and a wanted region that you want to plot
        from the csv 
    
    Parameters
    ----------
    csv : str
        The absolute or relative path of the csv file

    Returns
    -------
    None
    """
    df = pd.read_csv(csv)
    df = df[df['ID'] == region]
    df.plot(x='Datetime', y='Value', kind='line', title=region, figsize=(15,10))
    

plot_region_rescaled('../EXCEL_TO_CSV/all_data_long_rescaled.csv', 'ATHENS')
plot_region_rescaled('../EXCEL_TO_CSV/all_data_long_rescaled.csv', 'VOLOS')
plot_region_rescaled('../EXCEL_TO_CSV/all_data_long_rescaled.csv', 'THESSALONIKI')
plot_region_rescaled('../EXCEL_TO_CSV/all_data_long_rescaled.csv', 'KARDITSA')
plot_region_rescaled('../EXCEL_TO_CSV/all_data_long_rescaled.csv', 'LARISSA')
```

As we can see from the plots above there is a clear seasonality in these timeseries after the rescale of the data. We can see that gas consumption peaks in winter months which is
completely reasonable

```python
import plotly.express as px
import pandas as pd


def plot_all_regions(csv):
    """Gets a csv of data and plots all regions
        from the csv (must be in long format DEEPTSF)
    
    Parameters
    ----------
    csv : str
        The absolute or relative path of the csv file

    Returns
    -------
    None
    """
    df = pd.read_csv(csv)
    fig = px.line(df, x='Datetime', y='Value', color='ID', title='Stations Line Plots')

    # Show the plot
    fig.show()
    
    
```

### Plotting Rescaled 6H Data Resampled from DEEPTSF

As we can see there is not significant difference between 6H and 1H for ATHENS station timeseries
